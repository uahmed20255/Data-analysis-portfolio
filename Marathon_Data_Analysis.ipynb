```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates' : 'race_day',
                            'Event name' : 'race_name',
                            'Event distance/length' : 'race_distance',
                            'Athlete performance' : 'athlete_performance',
                            'Athlete gender' : 'athlete_gender',
                            'Athlete average speed' : 'average_speed',
                            'Athlete ID' : 'athlete_id'



                            })
```

```python
#difference in speed for the 50k , 50mi male to female
```

```python
df3.groupby(['race_distance', 'athlete_gender'])['average_speed'].mean()
```

```python
#what age groups are the best in the 50m race (20 + race min)
```

```python
df3.query('race_distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count > 19')
```

```python
#what age groups are the worst in the 50m race (10 + race min)(show 20)
```

```python
df3.query('race_distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count > 9')
```

```python
# seasons for the data


#sprint 3-5
#summer 6-8
#fall 9-11
#winter 12 -2
#split between 2 decimals
```

```python
df3['race_month']= df3 ['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_month'] = pd.to_numeric(df3['race_month'], errors='coerce')
df3['race_month'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x>5  else 'Spring'  if x > 2 else 'Winter')
```

```python
df3.head(25)
```

```python


# Now you can group by 'race_season'
df3.groupby('race_season')['average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
